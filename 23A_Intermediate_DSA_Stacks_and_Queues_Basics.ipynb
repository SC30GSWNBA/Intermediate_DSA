**Class Question1:** Parenthesis Balancing: String comprising of round brackets '(' , ')'. Return True if the string is balanced. For Example '( ( ) )' is balanced, '( ( ( (  ) ) ) )' is balanced, '( ( ( ) ) ) ( ) ) ) ) ( ( ) )' is not balanced.

**Hint:** Last coming in bracket '(' is the first one getting balanced. In the end if the stack is empty, then we can say that the string is balanced.

Stacks_n_Queues1.JPG

```python
# Code Implementation

def BalanceParenthesis(s):

  stack = []

  for bracket in s:

    if bracket == '(':

      stack.append(bracket)

    else:

      if len(stack) == 0:

        return False

      else:

        stack.pop()

  if len(stack) == 0:

    return True

  else:

    return False

print(BalanceParenthesis('(())'))
print(BalanceParenthesis('(())())('))
```

The above solution has **T.C. -> O(N)** and **S.C. -> O(N)**

```python
# Code Implementation

def BalanceParenthesis(s):

  cnt = 0

  for bracket in s:

    if bracket == '(':

      cnt += 1

    else:

      if cnt == 0:

        return False

      else:

        cnt -= 1

  if cnt == 0:

    return True

  else:

    return False

print(BalanceParenthesis('(())'))
print(BalanceParenthesis('(())())('))
```

The above solution has **T.C. -> O(N)** and **S.C. -> O(1)**

**Class Question1:** Parenthesis Balancing: Variation of the above problem with all kinds of brackets

    For Example - '([{]})' is not a balanced parenthesis

    But, '({[]})' is a valid parenthesis

```python
class Solution:
    def isValid(self, s: str) -> bool:
        
        char_list = []
        
        brackets = {'()', '[]', '{}'}
        
        for ch in s:

            if ch in '([{':
                
                char_list.append(ch)
                #print(char_list)
                

            elif not char_list or char_list.pop() + ch not in brackets:

                return False
            
        return not char_list

if __name__ == '__main__':
  X = Solution()
  print(X.isValid("()"))
  print(X.isValid("()[]{}"))
  print(X.isValid("(]"))
  print(X.isValid("([])"))
  print(X.isValid("([{]})"))
  print(X.isValid("({[]})"))
```

**Class Question2:** You are given 3 digits 1,2,3. You have to return Kth smallest number formed by these digits.

Queue3.JPG

```python
# Code Implementation

import collections

def KthNumber(digits, k):

  digits.sort()

  qu = collections.deque(digits)

  cnt = 0
  current_val = 0

  while cnt < k:

    current_val = qu.popleft()

    for digit in digits:

      qu.append(10*current_val + digit)

    cnt += 1

  return current_val

print(KthNumber([1,2,3], 7))
```
