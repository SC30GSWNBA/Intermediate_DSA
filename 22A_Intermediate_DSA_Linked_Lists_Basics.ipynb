**Class Question2:** Get the kth Node in a linked list

```python
def findkth(head, k):

  curr = head

  # For the kth node we move our pointer k-1 times

  for i in range(k-1):

    curr = curr.next

  return curr.data
```

```python
first_node = Node(1)
second_node = Node(2)
third_node = Node(3)
fourth_node = Node(4)
fifth_node = Node(5)
sixth_node = Node(6)
```

**Insertions:**

**Class Question3:** You are given a linked list, insert 5 at the beginning of linked list 1 -> 3 -> 4

```python
def insertatbegin(head, v):

  new_node = Node(v)

  new_node.next = head

  return new_node

nn = insertatbegin(first_node, 5)
```

```python
print_ll(nn)
```

**Class Question4:** You are given a linked list, insert 10 at the 4th position in the 5 -> 1 -> 2 -> 3 -> 4 -> 5 -> 6

```python
def insertatkth(head, v, k):

  if k == 1:

    return insertatbegin(head, v)

  curr = head

  for i in range(k-2):

    curr = curr.next

  temp = curr.next
  new_node = Node(v)
  curr.next = new_node
  new_node.next = temp
  return head

x = insertatkth(nn, 10, 4)
print_ll(x)
```

**Deletion**

**Class Question4:** You are given a linked list, delete element at 4th position in the 5 -> 1 -> 2 -> 10 -> 3 -> 4 -> 5 -> 6

```python
def deleteatkth(head, k):

  curr = head

  for i in range(k-2):

    curr = curr.next

  curr.next = curr.next.next

  return head

x = deleteatkth(nn, 4)
print_ll(x)
```

**Class Question5:** Find the length of a linked list

```python
def size(head):

  # return size of a linked list starting at head

  if head == None:

    return 0

  return 1 + size(head.next)

print(size(nn))
```
