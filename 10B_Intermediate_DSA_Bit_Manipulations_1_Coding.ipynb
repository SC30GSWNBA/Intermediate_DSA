**Q1. Single Number**

**Problem Description**

Given an array of integers A, every element appears twice except for one. Find that integer that occurs once.

**NOTE:** Your algorithm should have a linear runtime complexity. Could you implement it without using extra memory?

**Problem Constraints**

    2 <= |A| <= 2000000

    0 <= A[i] <= INTMAX

**Input Format**

The first and only argument of input contains an integer array A.

**Output Format**

Return a single integer denoting the single element.

**Example Input**

Input 1:

 A = [1, 2, 2, 3, 1]

Input 2:

 A = [1, 2, 2]

**Example Output**

Output 1:

 3

Output 2:

 1

**Example Explanation**

Explanation 1:

3 occurs once.

Explanation 2:

1 occurs once.

```python
class Solution:
  # param A : tuple of integers
  # return an integer
  def singleNumber(self, A):

    x = A[0]

    for i in range(1, len(A)):

      x = x ^ A[i]

    return x

X = Solution()

print(X.singleNumber([1, 2, 2, 3, 1]))
print(X.singleNumber([1, 2, 2]))
```

**Q2. Add Binary Strings**

**Problem Description**

Given two binary strings, return their sum (also a binary string).

**Example:**

a = "100"

b = "11"

Return a + b = "111".

```python
class Solution:
  # param A : string
  # param B : string
  # return a strings
  def addBinary(self, A, B):

    if len(A) >= len(B):

      B = '0'*(len(A)-len(B)) + B
            
    else:

      A = '0'*(len(A)-len(B)) + A

    final = ''

    flag = 0

    for i in range(len(A)-1, -1, -1):
            
      if A[i] == '0' and B[i] == '0' and flag == 0:
                
        flag = 0
        final = '0' + final
                
      elif A[i] == '0' and B[i] == '0' and flag == 1:
                
        flag = 0
        final = '1' + final
                
      elif A[i] == '0' and B[i] == '1' and flag == 0:
                
        flag = 0
        final = '1' + final
                
      elif A[i] == '1' and B[i] == '0' and flag == 0:
                
        flag = 0
        final = '1' + final
                
      elif A[i] == '1' and B[i] == '0' and flag == 1:
                
        flag = 1
        final = '0' + final
                
      elif A[i] == '0' and B[i] == '1' and flag == 1:
                
        flag = 1
        final = '0' + final
                
      elif A[i] == '1' and B[i] == '1' and flag == 0:
                
        flag = 1
        final = '0' + final
                
      else:
                
        flag = 1
        final = '1' + final
                
                
    if flag == 1:
            
      final = str(flag) + final

      return final
            
    else:
            
      return final

X = Solution()

print(X.addBinary("100","11"))
print(X.addBinary("1001","1101"))
```

**Q3. Number of 1 Bits**

**Problem Description**

Write a function that takes an integer and returns the number of 1 bits it has.

**Problem Constraints**

1 <= A <= 109

**Input Format**

First and only argument contains integer A

**Output Format**

Return an integer as the answer

**Example Input**

Input1:

11

**Example Output**

Output1:

3

**Example Explanation**

Explanation1:

11 is represented as 1011 in binary.

```python
# Solution-1

class Solution:
  # param A : integer
  # return an integer
  def numSetBits(self, A):

    count = 0

    for i in range(32):

      if A & (1<<i) != 0:

        count += 1

    return count

X = Solution()

print(X.numSetBits(11))
print(X.numSetBits(22))
```

```python
# Solution-2

class Solution:
  # param A : integer
  # return an integer
  def numSetBits(self, A):

    count = 0

    while A != 0:

      if A & 1 == 1:

        count += 1

      A = A >> 1

    return count

X = Solution()

print(X.numSetBits(11))
print(X.numSetBits(22))
```

**Q4. Interesting Array**

**Problem Description**

You have an array A with N elements. We have two types of operation available on this array :

    1. We can split an element B into two elements, C and D, such that B = C + D.

    2. We can merge two elements, P and Q, to one element, R, such that R = P ^ Q i.e., XOR of P and Q.

You have to determine whether it is possible to convert array A to size 1, containing a single element equal to 0 after several splits and/or merge?

**Problem Constraints**

    1 <= N <= 100000

    1 <= A[i] <= 106

**Input Format**

The first argument is an integer array A of size N.

**Output Format**

Return "Yes" if it is possible otherwise return "No".

**Example Input**

Input 1:

 A = [9, 17]

Input 2:

 A = [1]

**Example Output**

Output 1:

 Yes

Output 2:

 No

**Example Explanation**

Explanation 1:

Following is one possible sequence of operations - 

    1) Merge i.e 9 XOR 17 = 24  
    2) Split 24 into two parts each of size 12  
    3) Merge i.e 12 XOR 12 = 0  

As there is only 1 element i.e 0. So it is possible.

Explanation 2:

There is no possible way to make it 0.

**Hint:** 

    XOR of two even numbers is always even

    XOR of two odd is always even
    
    XOR of one even and one odd number is always odd

```python
class Solution:
  # param A : list of integers
  # return a strings
  def solve(self, A):

    cnt = 0

    for i in range(len(A)):
            
      if A[i]%2 == 1:
                
        cnt += 1
                
    if cnt%2 == 0:
            
      return "Yes"
            
    else:
            
      return "No"

X = Solution()

print(X.solve([9, 17]))
print(X.solve([1]))
```

**Q5. Reverse Bits**

**Problem Description**

Reverse the bits of an 32 bit unsigned integer A.

**Problem Constraints**

0 <= A <= 232

**Input Format**

First and only argument of input contains an integer A.

**Output Format**

Return a single unsigned integer denoting the decimal value of reversed bits.

**Example Input**

Input 1:

 0

Input 2:

 3

**Example Output**

Output 1:

 0

Output 2:

 3221225472

**Example Explanation**

Explanation 1:

    00000000000000000000000000000000    
    00000000000000000000000000000000

Explanation 2:

    00000000000000000000000000000011    
    11000000000000000000000000000000

```python
class Solution:
  # param A : unsigned integer
  # return an unsigned integer
  def reverse(self, A):

    result = 0
        
    for i in range(32):

      print(f'i = {i}')   
      temp = (A >> i) & 1

      print(f'temp: {temp}')
            
      result = result | (temp << (31 - i))

      print(f'result: {result}')
            
    return result

X = Solution()

print(X.reverse(0))
print('----------')
print(X.reverse(3))
```
