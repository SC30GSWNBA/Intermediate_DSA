**What is a Subarray?**

**Answer:** Contiguous part of an array and they are never empty.

For an array [1, 2, 3, 4], the total number of possible subarrays = 10

Total Subarrays for an array of N length = (N * (N+1)//2)

**What is a Subsequence?**

**Answer:** Sequence formed by deleting one or more elements from the array. The elements in a subsequence are present in the same order in the original array.

For Example,

    For an array [2, 1, 5, 4, 3, 1], [2, 5, 3, 1] is a valid subsequence, but [2, 3, 5, 1] is not a 
    valid subsequence because it is not in the same order as the original array.

    Every subarray is a subsequence, but every subsequence may or may not be a subarray.

    For an array [1, 2, 3], all possible subsequences: [1], [1,2], [1,2,3], [2], [2,3], [3], [1, 3], []

    For an array of size N, number of subsequences : 2^N

**What is a subset?**

**Answer:** Sequence formed by deleting one or more elements from the array. The elements in a subset are present in any order when compared to the original array.

For Example,

    For an array [2, 1, 5, 4, 3, 1], [2, 5, 3, 1] is a valid subset, and [2, 3, 5, 1] is 
    also a valid subset.

    Every subarray is a subset, but every subset may or may not be a subarray.

    Every subsequence is a subset, but every subset may or may not be a subseqeunce.

**Class Question1:** Given an array of N elements. Print all subsequences with sum = K ?

```python
# Code Implementation

def PossibleSubsequence(A, K):

  n = len(A)

  output = []

  for i in range(2**n):

    curr = 0

    curr_arr = []

    for j in range(n):

      if i & (1 << j) != 0:

        curr += A[j]

        curr_arr.append(A[j])

    if curr == K:

      output.append(curr_arr)

  return output

print(PossibleSubsequence([1, 2, 3, 4, 6], 6))
```

The above solution has a **T.C. -> O(2^N * N)** and **S.C. -> O(N^2)**

**Class Question2:** Given a array [-7, 2, 5, 4], find the sum of max value of all the subsequences of the array.

**Please Note:** For an array of size i (a1, a2, ........ ai), which is sorted in ascending order, the sum of the max value of all the subsequences of the array =

    = a1 x 2^0 + a2 x 2^1 + a3 x 2^2 + ............... ai x 2^(i-1)

```python
# Code Implementation:

def SumMaxSubsequence(A):

  A.sort()

  total = 0

  for i in range(len(A)):

    total += A[i] * (2**i)

  return total

print(SumMaxSubsequence([7, -1, 5]))
```

The above solution has a **T.C. -> O(NLogN)** and **S.C. -> O(1)**

**Class Question3:** Given a array [-7, 2, 5, 4], find the sum of min value of all the subsequences of the array.

```python
# Code Implementation:

def SumMaxSubsequence(A):

  A.sort(reverse = True)

  total = 0

  for i in range(len(A)):

    total += A[i] * (2**i)

  return total

print(SumMaxSubsequence([7, -1, 5]))
```

The above solution has a **T.C. -> O(NLogN)** and **S.C. -> O(1)**

**Class Question4:** Given two strings s and t, return true if s is a subsequence of t, or false otherwise.




```python
# Solution-1

class Solution:
  def isSubsequence(self, s: str, t: str) -> bool:
        
    n = len(t)
        
    for i in range(2**n):
            
      curr = ''
            
      for j in range(n):
                
        if i & (1 << j) != 0:
                    
          curr += t[j]
                    
      if curr == s:
                
        return True
            
    return False

if __name__ == '__main__':
  
  X = Solution()
  print(X.isSubsequence('abc', 'ahbgdc'))
  print(X.isSubsequence('axc', 'ahbgdc'))
```

The above solution has a **T.C. -> O(N^2)** and **S.C. -> O(1)**

```python
# Solution-2

class Solution:
  def isSubsequence(self, s: str, t: str) -> bool:
        
    i, j = 0, 0
        
    while i < len(s) and j < len(t):
            
      if s[i] == t[j]:
                
        i += 1
                
      j += 1
            
    if i == len(s):
            
      return True
        
    else:
            
      return False

if __name__ == '__main__':
  
  X = Solution()
  print(X.isSubsequence('abc', 'ahbgdc'))
  print(X.isSubsequence('axc', 'ahbgdc'))
```

The above solution has a **T.C. -> O(N)** and **S.C. -> O(1)**
