**Class Question1:** Given an array and a number K. Return True is there exists a pair (i, j), such that i and j are distinct indices and array[i] + array[j] == K. A = [2, 7, 11, 13]

```python
# Implementation - 1 - Using Hashmaps

def DistinctPair(A,K):

  check_set = set()

  for x in A:

    if K-x in A:

      return True

    else:

      check_set.add(x)

  return False

print(DistinctPair([2, 7, 11, 13], 15))
print(DistinctPair([2, 7, 11, 13], 20))
print(DistinctPair([2, 7, 11, 13], 10))
```

The above solution will have **T.C.: O(N)** and **S.C.: O(N)**

```python
# Implementation - 2 - Using Two Pointers

def DistinctPair(A,K):

  A.sort()

  i = 0
  j = len(A)-1

  while i < j:

    if A[i] + A[j] == K:

      return True

    elif A[i] + A[j] > K:

      j -= 1

    else:

      i += 1

  return False

print(DistinctPair([2, 7, 11, 13], 15))
print(DistinctPair([2, 7, 11, 13], 20))
print(DistinctPair([2, 7, 11, 13], 10))
print(DistinctPair([1, 5, 12, 14, 15, 16, 18, 50, 100], 34))
```

The above solution will have **T.C.: O(NLogN)** and **S.C.: O(1)**

**Class Question3:** Given an array of N elements. Calculate number of distinct elements in every window of size K. A = [6, 3, 7, 3, 8, 6, 9]

```python
# Implementation - Brute Force

def CountDistinctinWindow(A, K):

  ans = []

  n = len(A)

  for i in range(n-K+1):

    temp_dict = {}

    for j in range(i, i+K):

      if A[j] in temp_dict:

        temp_dict[A[j]] += 1

      else:

        temp_dict[A[j]] = 1

    ans.append(len(temp_dict))

  return ans

print(CountDistinctinWindow([6, 3, 7, 3, 8, 6, 9],3))
```

```python
# Implementation - Sliding Window Technique

def CountDistinctinWindow(A, K):

  ans = []

  freq = {}

  n = len(A)

  # Create Frequency table for the first window

  for i in range(K):

    if A[i] in freq:

      freq[A[i]] += 1

    else:

      freq[A[i]] = 1

  ans.append(len(freq))

  for i in range(K, n):

    if A[i] in freq:

      freq[A[i]] += 1

    else:

      freq[A[i]] = 1

    freq[A[i-K]] -= 1

    if freq[A[i-K]] == 0:

      del freq[A[i-K]]

    ans.append(len(freq))

  return ans


print(CountDistinctinWindow([6, 3, 7, 3, 8, 6, 9],3))
```

**Class Question4:** Given N array elements. Find the length of longest series/sequence of consecutive elements. A = [3, 1, 2, 100, 101]

```python
def LongestSequence(A):

  check_set = set(A)

  max_len = 0

  for num in A:

    if num-1 in check_set:

      continue

    else:

      curr_len = 0

      curr_ele = num

      while curr_ele in check_set:

        curr_ele += 1
        curr_len += 1

      max_len = max(curr_len, max_len)

  return max_len

print(LongestSequence([3, 1, 2, 100, 101]))
print(LongestSequence([3, 1, 2, 100, 101, 4, 102, 5, 6]))
```
