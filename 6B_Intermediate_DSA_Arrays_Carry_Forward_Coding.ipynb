**Q1. Special Subsequences "AG"**

**Problem Description**

You have given a string A having Uppercase English letters.

You have to find how many times subsequence "AG" is there in the given string.

**NOTE:** Return the answer modulo 109 + 7 as the answer can be very large.



**Problem Constraints**

1 <= length(A) <= 105



**Input Format**

First and only argument is a string A.



**Output Format**

Return an integer denoting the answer.



**Example Input**

Input 1:

 A = "ABCGAG"

Input 2:

 A = "GAB"


**Example Output**

Output 1:

 3

Output 2:

 0


**Example Explanation**

Explanation 1:

 Subsequence "AG" is 3 times in given string 

Explanation 2:

 There is no subsequence "AG" in the given string.

```python
# Solution 1

class Solution:
  # param A : string
  # return an integer
  def solve(self, A):

    n = len(A)

    cnt = 0

    for i in range(n):
      if A[i] == 'A':
        for j in range(i+1, n):
          if A[j] == 'G':
            cnt += 1

    return cnt

X = Solution()

print(X.solve("ABCGAG"))
print(X.solve("GAB"))
```

For the above problem, **T.C** -> O(N^2), S.C -> O(1)

```python
# Solution 2 - Suffix Sum

class Solution:
  # param A : string
  # return an integer
  def solve(self, A):

    n = len(A)

    if n == 1:

      return 0 # Length of the array is 1, so there would be 0 "12" pairs

    cnt = 0

    SS = [0] * n

    for i in range(n-1, -1, -1):

      if A[i] == 'G':

        cnt += 1
        SS[i] = cnt

      else:

        SS[i] = cnt

    #print(f'Suffix Sum: {SS}')

    cnt = 0

    for i in range(0, n-1):

      if A[i] == 'A':

        cnt += SS[i+1]

    return cnt

X = Solution()

print(X.solve("ABCGAG"))
print(X.solve("GAB"))
```

For the above problem, **T.C** -> O(N), **S.C** -> O(N)

```python
# Solution 3 - Carry Forward

class Solution:
  # param A : string
  # return an integer
  def solve(self, A):

    cnt_G = 0

    for i in range(len(A)):

      if A[i] == 'G':

        cnt_G += 1

    cnt_ans = 0

    for i in range(len(A)):

      if A[i] == 'A':

        cnt_ans += cnt_G

      if A[i] == 'G':

        cnt_G -= 1

    return cnt_ans

X = Solution()

print(X.solve("ABCGAG"))
print(X.solve("GAB"))
```

For the above problem, **T.C** -> O(N), **S.C** -> O(1)

**Q2. Closest MinMax**

**Problem Description**

Given an array A, find the size of the smallest subarray such that it contains at least one occurrence of the maximum value of the array and at least one occurrence of the minimum value of the array.



**Problem Constraints**

1 <= |A| <= 2000



**Input Format**

First and only argument is vector A



**Output Format**

Return the length of the smallest subarray which has at least one occurrence of minimum and maximum element of the array



**Example Input**

Input 1:

A = [1, 3]

Input 2:

A = [2]


**Example Output**

Output 1:

 2

Output 2:

 1


**Example Explanation**

Explanation 1:

 Only choice is to take both elements.

Explanation 2:

 Take the whole array.

```python
class Solution:
  # param A : list of integers
  # return an integer
  def solve(self, A):

    # Find max element in the array
    mx = max(A)

    # Find min element in the array
    mn = min(A)

    if mx == mn:

      return 1

    n = len(A)

    ans = len(A)

    # Keeping max as the starting element, find the occurence of the min element on the right of the max occurence, such that length of the subarray is minimum

    min_ind = 999999999999

    # Considering minimum index element is to the right of the maximum element

    for i in range(n-1, -1, -1):
      if A[i] == mx:
        ans = min(ans, min_ind-i+1)
      if A[i] == mn:
        min_ind = i

    # Considering minimum index element is to the left of the maximum element

    max_ind = 999999999999

    for i in range(n-1, -1, -1):
      if A[i] == mn:
        ans = min(ans, max_ind-i+1)
      if A[i] == mx:
        max_ind = i

    return ans


X = Solution()

print(X.solve([1, 3]))
print(X.solve([2]))
print(X.solve([2, 2, 6, 4, 5, 1, 5, 2, 6, 4, 3, 4, 1, 3]))
print(X.solve([8, 8, 8, 8, 8, 8]))
```

**Q3. Bulbs**

**Problem Description**

A wire connects N light bulbs.

Each bulb has a switch associated with it; however, due to faulty wiring, a switch also changes the state of all the bulbs to the right of the current bulb.

Given an initial state of all bulbs, find the minimum number of switches you have to press to turn on all the bulbs.

You can press the same switch multiple times.

**Note:** 0 represents the bulb is off and 1 represents the bulb is on.



**Problem Constraints**

1 <= N <= 5Ã—105

0 <= A[i] <= 1



**Input Format**

The first and the only argument contains an integer array A, of size N.



**Output Format**

Return an integer representing the minimum number of switches required.



**Example Input**

Input 1:

 A = [0, 1, 0, 1]

Input 2:

 A = [1, 1, 1, 1]


**Example Output**

Output 1:

 4

Output 2:

 0


**Example Explanation**

Explanation 1:

 press switch 0 : [1 0 1 0]

 press switch 1 : [1 1 0 1]

 press switch 2 : [1 1 1 0]

 press switch 3 : [1 1 1 1]
 
Explanation 2:

 There is no need to turn any switches as all the bulbs are already on.

```python
class Solution:
  # param A : list of integers
	# return an integer
	def bulbs(self, A):

		len_A = len(A)

		if len_A == 1 and A[0] == 0:

			return 1

		cnt = 0

		flag = 0

		for i in range(0, len_A):

			if A[i] == 0 and flag == 0:

				cnt += 1
				flag = 1

			elif A[i] == 1 and flag == 1:

				cnt += 1
				flag = 0

			else:

				pass

		return cnt

X = Solution()

print(X.bulbs([0, 1, 0, 1]))
print(X.bulbs([1, 0, 0, 1]))
print(X.bulbs([1, 0, 1, 0]))
print(X.bulbs([0, 0, 1, 1]))
print(X.bulbs([1]))
print(X.bulbs([0]))
```

**Q4. Amazing Subarrays**

You are given a string S, and you have to find all the amazing substrings of S.

An amazing Substring is one that starts with a vowel (a, e, i, o, u, A, E, I, O, U).

**Input**

Only argument given is string S.

**Output**

Return a single integer X mod 10003, here X is the number of Amazing Substrings in given the string.

**Constraints**

1 <= length(S) <= 1e6

S can have special characters

**Example**

Input

    ABEC

Output

    6

**Explanation**

    Amazing substrings of given string are :
    1. A
    2. AB
    3. ABE
    4. ABEC
    5. E
    6. EC
    here number of substrings are 6 and 6 % 10003 = 6.

```python
class Solution:
  # param A : string
  # return an integer
  def solve(self, A):

    A = A.lower()

    total = 0

    for i in range(len(A)):
            
      if A[i] == 'a' or A[i] == 'e' or A[i] == 'i' or A[i] == 'o' or A[i] == 'u':
                
        total += len(A)-i
        total = total%10003

    return total

X = Solution()

print(X.solve('ABEC'))
```

**Q5. Even Subarrays**

**Problem Description**

You are given an integer array A.

Decide whether it is possible to divide the array into one or more subarrays of even length such that the first and last element of all subarrays will be even.

Return "YES" if it is possible; otherwise, return "NO" (without quotes).



**Problem Constraints**

1 <= |A|, A[i] <= 106



**Input Format**

The first and the only input argument is an integer array, A.



**Output Format**

Return a string "YES" or "NO" denoting the answer.



**Example Input**

Input 1:

 A = [2, 4, 8, 6]

Input 2:

 A = [2, 4, 8, 7, 6]


**Example Output**

Output 1:

 "YES"

Output 2:

 "NO"


**Example Explanation**

Explanation 1:

 We can divide A into [2, 4] and [8, 6].
 
Explanation 2:

 There is no way to divide the array into even length subarrays.

```python
class Solution:
  # param A : list of integers
  # return a strings
  def solve(self, A):

    # If the array is not of even length then it can never be divided into subarrays of equal even lengths

    if len(A)%2 != 0:

      return 'NO'

    A_len = len(A)

    # An array is also a subarray of the array. For example - [2, 4, 6, 8] is a subarray of the array [2, 4, 6, 8]
    # Also checking if it has first and last elements as even

    if A_len > 1 and A[0]%2 == 0 and A[-1]%2 == 0:

      return 'YES'

    else:

      return 'NO'


X = Solution()

print(X.solve([2, 4, 8, 6]))
print(X.solve([2, 4, 8, 7, 6]))
```

**Q6. Leaders in an array**

**Problem Description**

Given an integer array A containing N distinct integers, you have to find all the leaders in array A.

An element is a leader if it is strictly greater than all the elements to its right side.

**NOTE:** The rightmost element is always a leader.



**Problem Constraints**

1 <= N <= 105

1 <= A[i] <= 108



**Input Format**

First and only argument is an integer array A.



**Output Format**

Return an integer array denoting all the leader elements of the array.

**NOTE:** Ordering in the output doesn't matter.



**Example Input**

Input 1:

 A = [16, 17, 4, 3, 5, 2]

Input 2:

 A = [1, 2]


**Example Output**

Output 1:

 [17, 2, 5]
 
Output 2:

 [2]


**Example Explanation**

Explanation 1:

 Element 17 is strictly greater than all the elements on the right side to it.

 Element 2 is strictly greater than all the elements on the right side to it.

 Element 5 is strictly greater than all the elements on the right side to it.

 So we will return this three elements i.e [17, 2, 5], we can also return [2, 5, 17] or [5, 2, 17] or any other ordering.

Explanation 2:

 Only 2 the rightmost element is the leader in the array.

```python
class Solution:
  # param A : list of integers
  # return a list of integers
  def solve(self, A):

    #cnt = 1

    n = len(A)

    output = [A[-1]]

    max_ele = A[-1]

    for i in range(n-2, -1, -1):

      if A[i] > max_ele:

        max_ele = A[i]
        output.append(max_ele)

        #cnt += 1

    #return cnt
    return output

X = Solution()

print(X.solve([16, 17, 4, 3, 5, 2]))
print(X.solve([1, 2]))
```
