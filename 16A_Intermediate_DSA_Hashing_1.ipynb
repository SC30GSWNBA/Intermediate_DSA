**Examples of HashMaps:**

> Store Population of all the Countries in the world

    Key: Country Name

    Population: Value

**Please Note:** 

    Key is unique for each and every <key, value> pair in Hash-Maps. 
    
    Every Key can have one of Multiple values.

    Keys are immutable in nature and can be either string, int, tuple.

    If a key is overwritten, it will replace/override the previous entry.

    Values can be any data structure. (List, Int, String, Tuple, Dictionary, Sets).

**Sets vs Dictionaries**

    Sets store only keys, dictionaries store key value pairs.

**Class Question1:** We are given an array of N elements. You have Q queries. Tell the frequencies of the element given in the query. array = [1, 1000, 10^9, 1000, 1, 2, 10^9, 2], queries = [1, 2, 3, 1000, 10001, 10^9]

```python
# Implementation

def CountFrequency(nums, queries):

  ele_freq = {}

  for i in range(len(nums)):

    if nums[i] in ele_freq:

      ele_freq[nums[i]] += 1

    else:

      ele_freq[nums[i]] = 1

  for x in queries:

    if x in ele_freq:

      pass
    
    else:

      ele_freq[x] = 0

  return ele_freq

print(CountFrequency([1, 1000, 1000000000, 1000, 1, 2, 1000000000, 2], [1, 2, 3, 1000, 10001, 1000000000]))
```

The above solution will have **T.C.: O(N)** and **S.C.: O(N)**

**Class Question3:** You are given an array of N elements. Find the count of distinct elements in the array. 

nums = [3, 5, 6, 5, 4]

```python
def CountDistinct(A):

  return len(set(A))

print(CountDistinct([3, 5, 6, 5, 4]))
```

**Class Question4**: Given N array elements. Check if there exists a subarray with sum = 0 ? nums = [2, 2, 1, -3, 4, 3, 1, -2, -3, 2]

```python
# Implementation

def SubarraySumZero(A):

  # Prefix element - pre is used to store sum
  pre = 0

  # The sum of the subarray is zero at the beginning
  store = {0}

  for e in A:

    pre += e

    # Checking if prefix sum is repeating itself
    if pre in store:

      return True

    store.add(pre)

  return False

print(SubarraySumZero([2, 2, 1, -3, 4, 3, 1, -2, -3, 2]))
```
