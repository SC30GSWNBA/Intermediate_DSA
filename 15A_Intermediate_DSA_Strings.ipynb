**What are Strings ?**

A String is an array of characters(numbers, symbols, alphabets) which is ordered.

**ASCII Values**

Every character(numbers, special characters, alphabets etc.) in a string has a 
unique number mapped to it. 

For Example 'a' to 'z' has ASCII values from 97 to 112. So, In Python, ord('a') = 97 and chr(97) = 'a'.

ASCII value is a mapping for characters in the form of a table.

    ASCII of 'a' = 97 = (1100001) in binary representation

    ASCII of 'A' = 65 = (1000001) in binary representation.

    So, there is a difference of 32 or 2^5 or 100000 in binary representation between 'a' and 'A'.
    In short, ord('a') - ord('A') = ord('b') - ord('B') = ............. = ord('z') - ord('Z') = 32

**Strings are Immutable**

Strings are immutable in Python and cannot be changed.

**Class Question1:** Given a string ('ScAlEr'), toggle('sCaLeR') the case of a string.

```python
# Solution-1 : Using ASCII Values

def ToggleCase(X):

  ans = ''

  for c in X:

    if ord(c) >= 97:

      ans += chr(ord(c)-32)

    else:

      ans += chr(ord(c)+32)

  return ans

print(ToggleCase('ScAlEr'))
print(ToggleCase('sCaLeR'))
```

```python
# Solution-2 : Using XOR to Toggle 6th bit

def ToggleCase(X):

  ans = ''

  for c in X:

    ans += chr(ord(c) ^ (1<<5))

  return ans

print(ToggleCase('ScAlEr'))
print(ToggleCase('sCaLeR'))
```

The above solution will have **T.C.: O(N)** and **S.C.: O(N)**

**Class Question2:** Given a string. Reverse a part of the string.

```python
def ReverseString(X, start, end):

  temp_list = [x for x in X]

  while start < end:

    temp_list[start], temp_list[end] = temp_list[end], temp_list[start]

    start += 1
    end -= 1

  #print(temp_list)

  return ''.join(temp_list)

print(ReverseString('INTRODUCTION', 2, 8))
```

The above solution will have **T.C.: O(N)** and **S.C.: O(N)**

**Class Question3:** Check if a string is Palindrome ?

```python
def CheckPalindrome(X):

  i = 0
  j = len(X)-1

  while i < j:

    if X[i] != X[j]:

      return False

    i += 1
    j -= 1

  return True

print(CheckPalindrome('NAMAN'))
print(CheckPalindrome('BAMAN'))
print(CheckPalindrome('ARORA'))
print(CheckPalindrome('MALAYALAM'))
```

The above solution will have **T.C.: O(N)** and **S.C.: O(1)**

**Class Question4:** Check if a subtring of a string is a palindrome.

```python
def CheckSubstring(X,i,j):

  while i < j:

    if X[i] != X[j]:

      return False

    i += 1
    j -= 1

  return True

print(CheckSubstring('NAMAN',1,3))
print(CheckSubstring('BAMAN',2,4))
print(CheckSubstring('ARORA',0,2))
print(CheckSubstring('MALAYALAM',2,6))
```

**Class Question5:** Given a string. Find the length of the longest palindromic substring.

```python
# Solution-1: Brute Force

def LongestPalindrome(X):

  ans = 0

  for i in range(len(X)):

    for j in range(i, len(X)):

      temp = X[i: j+1]

      if temp == temp[::-1]:

        if ans < len(temp):

          ans = len(temp)
          start = i
          end = j+1

  return X[start:end]

print(LongestPalindrome('BAMAN'))
```

**Centres in Strings:** In a sting 'ABCEDF', even length subtrings will have center in the middle of two middle characters. On, the other hand, odd length substrings will have center as the middle element in the substring. For Example:

    'ABC' has 3 characters as 'A','B','C', with 'B' as the center. Similarly, 'ABCED' will have
    characters 'A','B','C','E','D', with 'C' as the center.

    'AB' has 2 characters as 'A','B', with center between 'A' and 'B'. Similarly, 'ABCE' will 
    have characters 'A','B','C','E', with center between 'B' and 'C'.

**What are Comparators?**

**Answer:** Comparator replaces the comparison operation happening in the algorithm because the basis of sorting is comparison of 2 values on the basis of some parameter.

**How Comparators Work:** Supposedly we are trying to compare x and y on the basis of value, then

    - if x is smaller than y return -1

    - if x is greater than y return 1

    - if x is equal to y return 0

**Class Question6:** Sort array according to factors of the elements in an array. Given array, A = [12, 7, 6, 3, 1]. 

**Solve using comparators**

```python
# Code Implementation

# Importing comparator library

from functools import cmp_to_key

# Defining function to find total number of factors of a number

def CountFactors(n):

  count = 0

  for i in range(1, n+1):

    if n%i == 0:

      count += 1

  return count

# Defining function to compare number of factors of 2 numbers

def compare(a, b):

  n1 = CountFactors(a)

  n2 = CountFactors(b)

  if n1 < n2:

    return -1

  elif n1 > n2:

    return 1

  else:

    return 0

A = [12, 7, 6, 4, 1]
A.sort(key=cmp_to_key(compare))
print(A)
```

**In the above code,**

    'key' maps the values to new values based on which sorting happens, whereas,

    'comparator' changes the comparison operations happening while sorting. Since, the basis of 
    sorting is comparison, we directly change the comparison results.

**Class Question7:** Given an array A of non-negative integers, arrange them such that they form the largest number.

**Example Input**

Input 1:

 A = [3, 30, 34, 5, 9]

Input 2:

 A = [2, 3, 9, 0]

**Example Output**

Output 1:

 "9534330"
 
Output 2:

 "9320"

```python
# Code Implementation

def largestNumber(A):

  A = [str(x) for x in A]

  from functools import cmp_to_key

  def compare(a, b):

    if a + b > b + a:
		
      return -1
			
    if a + b < b + a:
		
      return 1
			
    return 0

  A.sort(key=cmp_to_key(compare))

  result = ''.join(A)

  if result=='0'*len(result):

    return '0'

  else:
            
    return result

print(largestNumber([3, 30, 34, 5, 9]))
print(largestNumber([2, 3, 9, 0]))
```

**Class Question8:** Given a string. Find the length of the longest palindromic substring. (Optimized Approach)

```python
def OptimizedLongestPalindrome(s):

  # Iterate on every character as center

  # Total n centres

  n = len(s)

  ans = 0

  for i in range(n):

    # s[i] is the center

    l = i-1

    r = i+1

    curr_len = 1

    while l >= 0 and r < n:

      if s[l] != s[r]:

        break

      curr_len += 2

      l -= 1

      r += 1

    ans = max(ans, curr_len)

  
  # Now considering centres as space between characters

  for i in range(n-1):

    l = i

    r = i+1

    curr_len = 0

    while l >= 0 and r <= n-1:

      if s[l] != s[r]:

        break

      curr_len += 2

      l -= 1

      r += 1

    ans = max(ans, curr_len)

  return ans

print(OptimizedLongestPalindrome('ababababab'))
```

**Class Question9:** Given a string. Find the longest palindromic substring.

```python
# Solution-1 - Using Recursion

class Solution:

    def longestPalindrome(self, s: str) -> str:

        def findpal(s,l,r):
            
            ls = len(s)
        
            while l>=0 and r<ls and s[l]==s[r]:
            
                l-=1
                r+=1
            
            return s[l+1:r]
        
        
        output = ""
        
        for i in range(len(s)):
            
            l,r = i,i #odd lengths
            
            check_palindrome = findpal(s,l,r)
            
            if len(output) <len(check_palindrome):
                
                output = check_palindrome
                
            l,r = i,i+1
            
            check_palindrome = findpal(s,l,r)
            
            if len(output) <len(check_palindrome):
                
                output = check_palindrome
                
        return output

X = Solution()

print(X.longestPalindrome('dbsbc'))
print(X.longestPalindrome('dbbc'))
print(X.longestPalindrome('ababababababab'))
print(X.longestPalindrome('babaddtattarrattatddetartrateedredividerb'))
```
