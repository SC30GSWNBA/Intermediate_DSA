**Class Question1:** What is the sum of the array elements?

arr = [12, 23, 34, 45, 56, 67, 78, 87]

```python
def SumOfArray(a):
  total = 0
  for i in a:
    total += i

  return total

arr = [12, 23, 34, 45, 56, 67, 78, 87]
```

```python
SumOfArray(arr)
```

For the above problem, **T.C** -> O(N), **S.C** -> O(1)

**Note:** We do not consider the input space in space complexity calculation. We only need to consider the auxiliary/extra variable space declared in the program.

**Class Question2:** Given N array elements, return the count of array elements which have atleast one element greater than themselves

arr = [2, 5, 1, 4, 8, 0, 8, 1, 3, 8]

```python
# Brute Force Approach

arr = [2, 5, 1, 4, 8, 0, 8, 1, 3, 8]

def CountElements(a):

  cnt = 0

  for i in range(len(a)):

    for j in range(len(a)):

      if a[i] < a[j]:
        
        cnt += 1
        break

  return cnt

CountElements(arr)
```

For the above problem, **T.C** -> O(N^2), **S.C** -> O(1)

**Optimized Algo:**



1.   Find max element
1.   Count total elements in the array
2.   Subtract the count of max elements in the array from the count of total elements in the array.

**Observation:** Max value won't have any value greater than itself



```python
arr = [2, 5, 1, 4, 8, 0, 8, 1, 3, 8]

def CountElements(a):

  mx = 0

  for i in a:

    mx = max(mx, i)

  #print(f'Max Element: {mx}')

  cnt_max = 0

  for i in a:

    if i == mx:

      cnt_max += 1

  return len(a)-cnt_max


CountElements(arr)
```

For the above problem, **T.C** -> O(N), **S.C** -> O(1)

**Class Question3:** Given an array, find if an element exists in an array?

arr = [1, 5, 7, 2, 6, 4, 3]

```python
# Approach 1

arr = [1, 5, 7, 2, 6, 4, 3]

def FindElement(a, k):

  return k in a

print(FindElement(arr, 5))
print(FindElement(arr, 10))
```

For the above problem, **T.C** -> O(N), **S.C** -> O(1)

```python
# Approach 2

arr = [1, 5, 7, 2, 6, 4, 3]

def FindElement(a, k):

  return k in a

print(FindElement(set(arr), 5))
print(FindElement(set(arr), 10))
```

For the above problem, **T.C** -> O(1), **S.C** -> O(N), because we are creating a new set inside print function

**Class Question 4:** Reverse a given array

```python
# Approach 1

def ReverseArray(arr):

  result = []

  for i in range(len(arr)-1, -1, -1):

    result.append(arr[i])
                  
  return result

print(ReverseArray([1, 5, 7, 2, 6, 4, 3]))
print(ReverseArray([1]))
print(ReverseArray([1, 5, 7, 2, 6, 4]))
```

For the above problem, **T.C** -> O(N), **S.C** -> O(N)

```python
# Approach 2

def ReverseArray(arr):

  if len(arr) == 1:
    return arr

  i = 0
  j = len(arr)-1

  while i < j:
    arr[i],arr[j] = arr[j],arr[i]
    i += 1
    j -= 1

  return arr

print(ReverseArray([1, 5, 7, 2, 6, 4, 3]))
print(ReverseArray([1]))
print(ReverseArray([1, 5, 7, 2, 6, 4]))
```

For the above problem, **T.C** -> O(N), **S.C** -> O(1)

```python
# Approach 3

arr = [1, 5, 7, 2, 6, 4, 3]

print(arr[::-1])
```

For the above problem, **T.C** -> O(N), **S.C** -> O(N), because slicing creates a new list and then appends the data in that list before returning the result. So, both T.C and S.C are O(N)

**Class Question 5:** Reverse a given array from index i to index j

arr = [1, 5, 7, 2, 6, 4, 3]

```python
def ReverseSubArray(arr, i, j):

  if len(arr) == 1:
    return arr

  while i < j:
    arr[i],arr[j] = arr[j],arr[i]
    i += 1
    j -= 1

  return arr

print(ReverseSubArray([1, 5, 7, 2, 6, 4, 3], 1, 5))
print(ReverseSubArray([1, 5, 7, 2, 6, 4], 1, 4))
```
