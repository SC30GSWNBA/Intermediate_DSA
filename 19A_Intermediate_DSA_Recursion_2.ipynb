**Class Question1:** Find the sum of all digits of a non-negative number using recursion.

```python
# Coding Implementation

def SumofDigits(num):

  if num >= 0 and num < 10:

    return num

  return (num%10) + SumofDigits(num//10)

print(SumofDigits(9879))
```

**Power Function in Recursion**

a ^ n = a x a x a x a ..... ntimes

```python
# Code Implementation

def PowerFunction(a, n):

  if n == 1:

    return a

  return a * PowerFunction(a, n-1)

print(PowerFunction(2, 10))
```

The above solution has a **T.C. -> O(N)** and **S.C. -> O(N)**

```python
# Code Implementation

def PowerFunctionOptimized(a, n):

  if n == 1:

    return a

  if n == 0:

    return 1

  if n%2 == 0:

    return PowerFunctionOptimized(a, n//2) * PowerFunctionOptimized(a, n//2)

  else:

    return a * PowerFunctionOptimized(a, (n-1)//2) * PowerFunctionOptimized(a, (n-1)//2)

print(PowerFunctionOptimized(2, 10))
```

The above solution has a **T.C. -> O(LogN)** and **S.C. -> O(N)**
